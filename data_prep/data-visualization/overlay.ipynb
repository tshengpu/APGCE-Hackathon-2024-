```python
import numpy as np
import pandas as pd
from PIL import Image, ImageFilter
from matplotlib import pyplot as plt
import os
import cv2
```

# Raw Images

```python
print(f"Available selection: {os.listdir('../../data/raw_fault')}")
```

```python
code = '1004'

img = cv2.imread(f'../../data/raw_seismic/seismic-{code}.png')
fault_mask = cv2.imread(f'../../data/raw_fault/fault-{code}.png')

overlay = cv2.addWeighted(img, 0.5, fault_mask, 0.5, 0)
fig, ax = plt.subplots(figsize=(13,10))
ax.imshow(overlay)
#cv2.imwrite(f"overlay_img_fault{code}.png", overlay)
```
