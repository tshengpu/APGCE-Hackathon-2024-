```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot for Accuracy

df = pd.read_csv(r'../pally/model_accuracy.csv')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Epoch', y='Training Accuracy', data=df, label='Training Accuracy', marker='o')
sns.lineplot(x='Epoch', y='Validation Accuracy', data=df, label='Validation Accuracy', marker='o')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Plot for Loss

df = pd.read_csv(r'../pally/model_accuracy.csv')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Epoch', y='Training Loss', data=df, label='Training Loss', marker='o')
sns.lineplot(x='Epoch', y='Validation Loss', data=df, label='Validation Loss', marker='o')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
